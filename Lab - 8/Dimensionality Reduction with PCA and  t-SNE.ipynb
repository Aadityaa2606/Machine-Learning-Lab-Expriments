# Objective
To perform dimensionality reduction with PCA and t-SNE on a custom dataset

# Dataset import

```python
import pandas as pd
from sklearn.decomposition import PCA
import plotly.express as px


df = pd.read_csv("data.csv")
```

```python
df.head()
```

Fitting PCA

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop(['database name original','contributing team','database name','filename','original filename'], axis=1)
y = df['digit']

scaler = StandardScaler()
X_norm = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_norm, y, random_state=13, test_size=0.25, shuffle=True
)

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)

pca.score(X_test)
```

# PCA visualization

```python

fig = px.scatter(x=X_train_pca[:, 0], y=X_train_pca[:, 1], color=y_train)
fig.update_layout(
    title="PCA visualization of Customer dataset",
    xaxis_title="First Principal Component",
    yaxis_title="Second Principal Component",
)
fig.show()
```

### t-SNE reduction

```python
# t-SNE fitting
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2,perplexity=40, random_state=42)
X_train_tsne = tsne.fit_transform(X_train)

tsne.kl_divergence_
```

### t-SNE visualization

```python
fig = px.scatter(x=X_train_tsne[:, 0], y=X_train_tsne[:, 1], color=y_train)
fig.update_layout(
    title="t-SNE visualization of Customer Churn dataset",
    xaxis_title="First t-SNE",
    yaxis_title="Second t-SNE",
)
fig.show()
```

# Conclusion
t-SNE, a popular dimensionality reduction technique that can visualize high-dimensional non-linear data in a low-dimensional space.
